# N=3 A = [0,2]

```python
p = np.linspace(0, 10, 100)
points = []
for i in p:
    H = ising_model_hamiltonian(3,1,i,0)
    #print(H)
    E, psi = H.groundstate()
    rhoa = Q.ptrace(psi,[0,2])
    #print(rhoa)
    eigenvals = rhoa.eigenenergies()
    threshold1 = abs(eigenvals) < 10**(-8)
    eigenvals[threshold1] = 0
    #print(eigenvals)
    eigenvalscleaned = [num for num in eigenvals if num != 0]
    #print(eigenvalscleaned)
    vnentropylist = []
    for l in eigenvalscleaned:
        vnentropylist.append(-l*np.log(l))
        #print(vnentropylist)
        vnentropy = sum(vnentropylist)
        #print(vnentropy)
    points.append(vnentropy)

fig, ax = plt.subplots(figsize=(8,8))
plt.plot(p, points, linestyle = "none", marker = "o")
# label the axes
ax.set_xlabel("h value")
ax.set_ylabel("Von Neumann Entropy")

# set the tick mark parameters
ax.minorticks_on()
ax.tick_params(which='both',direction="in",width=1)
ax.tick_params(which='major',length=7)
ax.tick_params(which='minor',length=3.5)
ax.xaxis.set_ticks_position('both')
ax.yaxis.set_ticks_position('both')
plt.show()
```
